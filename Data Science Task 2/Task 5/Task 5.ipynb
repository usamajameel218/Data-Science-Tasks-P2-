```python
# import liberaries
import streamlit as st
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_csv("D:\Internship Work\Data Sceince Task 2\Task 5\Global Superstore.csv", encoding='latin1')
```

```python
# Data Cleaning
# 1. Drop duplicate rows
df.drop_duplicates(inplace=True)
```

```python
# Remove rows with missing critical values
df.dropna(subset=["Region", "Category", "Sub-Category", "Sales", "Profit", "Customer Name"], inplace=True)
```

```python
# Convert date columns to datetime (if present)
if "Order Date" in df.columns:
    df["Order Date"] = pd.to_datetime(df["Order Date"], errors="coerce")
if "Ship Date" in df.columns:
    df["Ship Date"] = pd.to_datetime(df["Ship Date"], errors="coerce")
```

```python
# Ensure Sales and Profit are numeric
df["Sales"] = pd.to_numeric(df["Sales"], errors="coerce")
df["Profit"] = pd.to_numeric(df["Profit"], errors="coerce")
```

```python
# Drop rows with invalid numeric conversions
df.dropna(subset=["Sales", "Profit"], inplace=True)
```

```python
# Reset index after cleaning
df.reset_index(drop=True, inplace=True)

df.head()
```
